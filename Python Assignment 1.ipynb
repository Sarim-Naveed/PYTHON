## Q1. Variable
A variable in Python is a name used to store a value in memory.

```python
age = 20
print(age)
```

## Q2. Rules of Variable Naming
1. Must start with a letter or underscore.
2. Cannot start with a number.
3. No spaces allowed.

Correct Example: student_name  
Incorrect Example: 1name

## Q3. Integer Data Type
An integer is a whole number without a decimal point.

```python
num = 10
print(num)
```

## Q4. Float Data Type
A float is a number with a decimal point.

```python
price = 99.5
print(price)
```

## Q5. String Data Type
A string is a sequence of characters.

```python
name1 = "Ali"
name2 = 'Ahmed'
print(name1, name2)
```

## Q6. Boolean Data Type
Boolean data type has two values: True and False.


```python
is_student = True
print(is_student)
```

## Q7. type() Function
The type() function is used to check the data type of a variable.


```python
x = 5
print(type(x))
```

## Q8. input() Function
The input() function is used to take input from the user.
Default data type of input is string.


```python
name = input("Enter your name: ")
print(name)
print(type(name))
```

## Q9. Type Casting
Type casting means converting one data type into another.


```python
num_str = "10"
num_int = int(num_str)
print(num_int)
```
